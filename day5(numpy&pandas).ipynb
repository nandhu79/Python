```python
#samples data representation 
months = ['jan','feb','mar','apr','may','jun']
traffic = [150, 2, 180, 220, 250, 310]

#craete a line plot
plt.plot(months, traffic)

#Add labels and title 
plt.xlabel('Month')
plt.ylabel('Monthly Taffic')
plt.title('Monthly Website Traffic')

#Display the plot
plt.show()
```

```python
#samples data representation 
months = ['jan','feb','mar','apr','may','jun']
product_a_revenue = [45,33, 56,44,78,98,]
product_b_revenue = [35,40,35,67,54,56]

#craete a line plot
plt.plot(months, product_a_revenue, marker = 'o', linestyle = '-', color= 'blue', label = 'Product A')

#create a line plot for product b
plt.plot(months, product_b_revenue, marker = 's', linestyle = '-', color= 'purple', label = 'Product B')

#Add labels and title 
plt.xlabel('Month')
plt.ylabel('Monthly Taffic')
plt.title('Monthly Website Traffic')

#display a legend to differentiate between Product A and product B
plt.legend()

#Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

winnervslooser = ['Gujarat Titans', 'Rajasthan Royals', 'Gujarat Titans', 'Punjab Kings',
                  'Mumbai Indians', 'Delhi Daredevils', 'Rajasthan Royals', 
                  'Kolkata Knight Riders', 'Royal Challengers Bangalore', 'Delhi Daredevils']


team_counts = Counter(winnervslooser)

teams = list(team_counts.keys())
values = list(team_counts.values())


plt.bar(teams, values, color=['pink', 'green', 'black', 'blue', 'red', 'yellow', 'orange', 'purple'])

plt.xlabel('                                   Number of Wins')
plt.ylabel('Teams')
plt.title('Number of Wins by Each Team')

# Show plot
plt.show()
```
