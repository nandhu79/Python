```python
#Histograms

import matplotlib.pyplot as plt
import numpy as np

#Generate random data for histogram
data = np.random.random_sample(5)

list1  = [1,2,1,-1,1,2,3,-6,3,4,6,7]

plt.hist(data, color = 'red',edgecolor = 'yellow', bins = 5)

plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt

quaters = ['Q1','Q2','Q3','Q4']
electronics = [10000,20000,30000,40000]
clothing = [5000,400,7000,1000]
home_appliances  =[30000,7000,2000,3000]

plt.figure(figsize = (10,6))
plt.stackplot(quaters,electronics,clothing,home_appliances,labels = ['Electronics','Clothing','HomeAppliances'])

plt.xlabel('Quaters')
plt.ylabel('Values')
plt.title('Sales Data')

plt.grid()

# plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x,y)
z = np.sin(np.sqrt(x**2 + y**2 + (z/2)))
           
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(x,y,z, c= z, cmap = 'viridis')
ax.set_xlabel('X Axis')
ax.set_ylabel('Y Axis')
ax.set_zlabel('Z Axis')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x  = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

fig, axs = plt.subplots(2,2,figsize=(10,8))

axs[0,0].plot(x, y1, 'r')
axs[0,0].set_title('Sine Wave')
axs[0,1].plot(x, y2, 'b')
axs[0,1].set_title('Cosine Wave')
axs[1,0].plot(x, y3, 'g')
axs[1,0].set_title('Tan Wave')
axs[1,1].plot(x, y4, 'm')
axs[1,1].set_title('Exponential Decay')

plt.tight_layout()
plt.show()
```
