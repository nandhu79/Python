```python
import matplotlib.pyplot as plt

# Data
per_year = [2026, 2027, 2028, 2029]
salary_hike_data = [10, 20, 30, 40]
salary_hike_devops = [20, 30, 40, 10]
salary_hike_developers = [50, 30, 20, 60]

# Plotting the data
plt.plot(per_year, salary_hike_data, marker='o', linestyle='-', color='pink', label='Data Engineers')
plt.plot(per_year, salary_hike_devops, marker='s', linestyle='-', color='purple', label='DevOps')
plt.plot(per_year, salary_hike_developers, marker='p', linestyle='-', color='blue', label='Software Developers')

# Labels and title
plt.xlabel('Year')
plt.ylabel('Salary Hike (%)')
plt.title('Salary Hike of Employees')

# Display legend
plt.legend()

# Show the plot
plt.show()
```
