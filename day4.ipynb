```python
#an iterator is an object tht allows you to traverse through all the elements of a collection one at a time
```

```python
class MyNumbers:
    def __iter__(self):
        self.num = 1
        return self
    def __next__(self):
        if self.num <= 5:
            x = self.num
            self.num += 1
            return x
        else:
            raise StopIteration
myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
    print(x)
            
```

```python
#Generator: a generator is a sepcial type of iterator that is defined using a function and the 'yield' keyword

def my_generator():
  for i in range(1,6):
      yield i
gen = my_generator()

for value in gen:
    print(value)
```

```python
#iterator vs generator 
"""iterator: requires class with __iter__() and __next__()
Generator : Uses Functions with keyword
Iterator : more code, manual implementation
Generator : Simple and conclude implementation
Iterator: Slightly more control
Generator : More memory effect
```

```python
def my_generator(start, end):
    while start < end:
        yield start
        start += 1

gen = my_generator(0, 5)
for number in gen:
    print(number)
```

```python
class MyIterator:
    def __init__(self, start, end):
        self.current = start
        self.end = end

    def __iter__(self):
        return self

    def __next__(self):
        if self.current >= self.end:
            raise StopIteration
        self.current += 1
        return self.current - 1

my_iter = MyIterator(0, 5)
for number in my_iter:
    print(number)
```
