```python
#create a directory put the file inside a director ,write data inside the file,read the file and delete the directory
import os
import shutil

directory_name = "temp_directory"
file_name = "example.txt"
file_content = "Hello, Python! File handling is versatile."
file_path = os.path.join(directory_name, file_name)

try:
    os.makedirs(directory_name, exist_ok=True)
    print(f"Directory '{directory_name}' created.")

    
    with open(file_path, "w") as file:
        file.write(file_content)
    print(f"Data written to file '{file_path}'.")

  
    with open(file_path, "r") as file:
        read_content = file.read()
    print(f"Content read from file: '{read_content}'")

    os.remove(file_path)
    print(f"File '{file_name}' deleted.")

    os.rmdir(directory_name)
    print(f"Directory '{directory_name}' deleted.")

except FileExistsError:
    print(f"Error: Directory '{directory_name}' already exists.")
except FileNotFoundError:
    print(f"Error: File '{file_path}' not found.")
except OSError as e:
    print(f"An OS error occurred: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

```
