```python
#Matrix Transpose

matrix = [[1, 2, 3],
          [3, 4, 5],
          [6, 7, 8]
         ]
rows = len(matrix)
col = len(matrix[0])

transpose = []

for i in range(col):
    new_matrix=[]
    for j in range(rows):
        new_matrix.append(matrix[j][i])
    transpose.append(new_matrix)

print("Original matrix:")
for col in matrix:
    print(col)

print("Transpose matrix")
for col in transpose:
    print(col)
    
        
        
         
```

```python
#matrix transpose with function

def transpose_matrix(matrix):
    rows = len(matrix)
    col = len(matrix[0])

    transpose= []

    for i in range(col):
        new_col = []
        for j in range(rows):
            new_col.append(matrix[j][i]) 
        transpose.append(new_col)

    return transpose       

matrix = [[3, 4, 5],
          [4, 5, 6],
          [0, 9, 8]]  

result =  transpose_matrix(matrix)

print("Transpose matrix")
for rows in result:
    print(rows)
        
```
