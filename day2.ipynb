```python
class Sports:
    def __init__(self, model,colour, cost):
        self.__colour = colour
        self.__model = model
        self.__cost = cost
    def getcolour(self):
        return self.__colour
    def setcolour(self, colour):
        self.__colour = colour
    def getcost(self):
        return self.__cost
    def setcost(self, cost):
        self.__cost = cost
    def getmodel(self):
        return self.__model
    def setmodel(self, model):
        self.__model = model
class shoes(Sports):
    def __init__(self, model, colour, cost,brand):
        super().__init__(model,colour, cost)
        self.__brand = brand
sh1 = shoes("Shoes","tsedr", "black", "nik")
print(sh1.getmodel())
            
```

```python
#exception handling without using
import sys
lst = ['b',0,2]
for entry in lst:
    print("value is :", entry)
    k = 1/int(entry)
```

```python
#exception handling using
import sys
lst = ['b', 0 , 2]
for entry in lst:
    try:
        print("value is:", entry)
        k = 1/int(entry)
    except:
        print("oops", sys.exc_info(),"occured")
        print("next entry")
        print("*************************")
print("the reciprocal of", entry, "is", k)        
              
        
```

```python
import sys
lst = ['b',0,2]

for entry in lst:
    try:
        print("+++++++++++++++++++")
        print("Value is:",entry)
        k = 1/int(entry)

    except(ValueError):
        print("This is the valueError")
    except(ZeroDivisionError):
        print("ZeroDivisionError")
    except:
        print("Other error")
print("The reciprocal of", entry, "is", k)        
```

```python
#calculator with exception
def calculator():
   while True:
       try:
           
           num1 = float(input("Enter the first number: "))
           num2 = float(input("Enter the second number: "))
           
           operator = input("Enter an operator (+, -, *, /): ")
           if operator == "+":
               result = num1 + num2
           elif operator == "-":
               result = num1 - num2
           elif operator == "*":
               result = num1 * num2
           elif operator == "/":
               try:
                   result = num1 / num2
               except ZeroDivisionError:
                   print("Error: Division by zero is not allowed.")
                   continue
           else:
               print("Invalid operator. Please try again.")
               continue
           print(f"The result is: {result}")
       except ValueError:
           print("Invalid input. Please enter numeric values.")
       choice = input("Do u want to continue (y/n): ").lower()
       if choice != 'y':
           print("Exiting the calculator. Goodbye!")
           break
calculator()
```

```python
#calculator without exception
def calculator():
    a = float(input("Enter the first number: "))
    b = float(input("Enter the second number: "))
    op = input("Enter the operator (+, -, *, /): ")
    
    result = None 
    
    if op == "+":
        result = a + b
    elif op == "-":
        result = a - b
    elif op == "/":
        if b != 0: 
            result = a / b
        else:
            print("Error: Division by zero.")
    elif op == "*":
        result = a * b
    else:
        print("Invalid operator. Please try again.")
    

    if result is not None:
        print(f"The result is {result}")

calculator()
```
